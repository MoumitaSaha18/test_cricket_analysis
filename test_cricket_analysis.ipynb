**Meaning of the each Column**

1. Player: the name of the players who played test cricket.
<br>
2. Span: the career length of the players.
<br>
3. Mat: Number of matches played by the players.
<br> 
4. Inns: Total innings played by the players.
<br>
5. Balls: The number of balls bowled by the players. Overs are more traditional but balls are a more useful statistic becausse the number of balls per over has verified historically.
<br>
6. Runs: Number of runs acknowledged.
<br>
7. Wkts: Total wickets earned by the players.
<br>
8. BBI: Best bowling by the players.
<br>
9. BBM: Best bowling in the matches played by the players.
<br>
10. Ave: The average number of runs acknowledged per wicket taken. (Ave = Runs/W)
<br>
11. Econ: The average number of runs of balls acknowledged per over. (Econ = runs/Overs bowled).
<br>
12. SR = the average number of balls bowled per wicket taken. (SR = Balls/W)
<br>
13. 5: the number of innings in which the bowler took atleast five wickets.
<br>
There are four wickets in an innings, the number of innings in which the bowler took exactly four wickets, is sometimes recorded alongside five wickets.
<br>
14. 10: the number of matches in which the bowler took at least ten wickets;
recorded for test and first-class matches.

**Checking for missing values and data types of the each columns of the dataset.**

```python
# Showing The Data Types
df.dtypes
```

```python
# Showing the detailed overview of the dataset.
print(df.info())

#Checking if there any null value in the columns.
print('Checking if there are any null value in the columns:')
df.isnull()
```

```python
# Drop a row if it has atleast 1 null value in any of it's columns.
new_df = df.dropna(axis = 0, how = 'any')

#display the new dataset, without any null value of any columns.

display(new_df.head(5))
print("Old data set length:", len(df))
print("new data set length:", len(new_df))

#Count the rows that has at least 1 null values in any of it's columns.
print("Number of rows with at least 1 null Value: ", (len(df)-len(new_df)))

# Count non-null values of each columns
df.count()
```

```python
# Total Missing Value Count
display(df.isnull().sum())
```

**Renaming the column names appropriately**

```python
# again printing the column names to rename some of the column names 
print("Columns of the Dataset: ", df.columns)
```

```python
# Renamimng the columns' names which are a little bit hard to understand
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'Wkts': 'Wickets',
                        'BBI': 'Best_Bowling/Innings',
                        'BBM': 'Best_Bowling/Match',
                        'Ave': 'Average',
                        'Econ': 'Economy_Rate',
                        'SR': 'Strike_Rate',
                        '5': '5_Wickets',
                        '10': '10_Wickets'})

display(df.head())
```

**Remove a column from the dataframe**

```python
# Remove 10_Wickets columns as it has some null values in the rows.
df.drop('Best_Bowling/Match', axis=1, inplace=True)

display(df.head())
```
